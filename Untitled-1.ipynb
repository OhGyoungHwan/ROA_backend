```python
import pytesseract
import cv2 
import matplotlib.pyplot as plt
import imutils
import re
import requests
import numpy as np
from imutils.perspective import four_point_transform


def closing(image):
    kernel = np.ones((4,4),np.uint8)
    result = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
    return result


url = 'https://upload3.inven.co.kr/upload/2022/12/27/bbs/i14255462832.jpg?MW=800'
path = './in3.png'
image = cv2.imread(path)

image_nparray = np.asarray(bytearray(requests.get(url).content), dtype=np.uint8)
org_image = cv2.imdecode(image_nparray, cv2.IMREAD_COLOR) 

rgb_image = cv2.cvtColor(org_image, cv2.COLOR_BGR2HSV)

gray = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2GRAY)


# RGB이미지를 흑백이미지로 변환
binary_image = cv2.threshold(rgb_image[:,:,2],0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]

removenoise_image = cv2.medianBlur(binary_image,ksize=1)
plt.imshow(binary_image)
text = pytesseract.image_to_string(binary_image, lang='kor', config= "--psm 4")
text = text.replace("7|","기")

print(text)
```

```python
import pandas as pd
df = pd.read_csv('./data/csv/diablo2option_temp.csv')
dictmask = {}
f = open("./data/txt/diablo2compiler.txt",'w',encoding="UTF_8")
for i in df.종류.values.tolist():
    if pd.isnull(i) != False:
        continue
    for j in ['매직','레어','크래프트세이브티','크래프트힛파워','크래프트블러드','크래프트캐스터']:
        mask = df[(df['종류'] == i) & (df['희귀도'] == j)]
        if (len(mask.index) == 0):
            continue
        compiler =  mask.dropna(axis=1).columns.values.tolist()
        dictmask[i+j] = '|'.join(compiler[2:])
f.write(str(dictmask))
f.close()

        
```
